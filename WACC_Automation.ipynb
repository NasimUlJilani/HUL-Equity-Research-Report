```python
import yfinance as yf
import numpy as np
import pandas as pd
rf = float(input("Risk Free Rate (%): "))/100
rm = float(input("Market Returns (%): "))/100
ERP = rm-rf
print(f"ERP: {ERP*100:.2f}%")
kd_pt = float(input("Pre-Tax Cost of Debt (%): "))/100
mtr = float(input("Marginal Tax Rate (%): "))/100


```

```python
benchmark_ticker = input("Benchmark Ticker: ")
n_peers = int(input("Number of Peers: "))
start_date = input("Start Date (YYYY-MM-DD): ")
end_date = input("End Date (YYYY-MM-DD): ")
```

```python
companies = {}

name = input("\nTarget Company Name: ")
companies[name] = {
    "ticker": input(f"{name} Ticker: "),
    "mcap": float(input(f"{name} Market Capitalisation: ")),
    "debt": float(input(f"{name} Total Debt: "))
}
```

```python
for i in range(1,n_peers+1):
  name = input(f"\nPeer Company {i} Name: ")
  companies[name] = {
    "ticker": input(f"{name} Ticker: "),
    "mcap": float(input(f"{name} Market Capitalisation: ")),
    "debt": float(input(f"{name} Total Debt: "))
}

print("\nDone")
print(companies)
```

```python
tickers = [info["ticker"] for companies in companies.values()]
tickers.append(benchmark_ticker)
```

```python
tickers = [company_data["ticker"] for company_data in companies.values()]
tickers.append(benchmark_ticker)
data = yf.download(tickers,start = start_date,end = end_date)
```

```python
data.head(20)
```

```python
close_prices = data["Close"]

benchmark_close = close_prices[benchmark_ticker].resample("ME").last()
benchmark_ret = benchmark_close.pct_change().dropna()

for name,info in companies.items():
  info["close"] = close_prices[info["ticker"]].resample("ME").last()
  info["returns"] = info["close"].pct_change().dropna()

print(benchmark_close.head())
print(benchmark_ret.head())
```

```python
raw_betas = {}
for company,info in companies.items():

  aligned_returns = pd.concat([benchmark_ret, info["returns"]], axis=1).dropna()


  aligned_benchmark_ret = aligned_returns.iloc[:, 0]
  aligned_company_ret = aligned_returns.iloc[:, 1]

  raw_betas[company] = np.polyfit(aligned_benchmark_ret, aligned_company_ret, 1)[0]

  print(f"{company} Raw Beta: {raw_betas[company]:.4f}")
```

```python

w_company = 0.75
w_market = 0.25

market_beta = 1

adjusted_beta = {}

for company,raw_beta in raw_betas.items():
    adjusted_beta[company] = round((w_company*raw_beta)+(w_market*market_beta),4)

print(adjusted_beta)
```

```python
de_ratio = {}
for company, info in companies.items():
  de_ratio[company] = info["debt"]/info["mcap"]
  print(f"{company} Debt/Equity ratio: {de_ratio[company]:.4f}")
```

```python
unlvr_beta = {}

for company, info in companies.items():
  # Ensure we use the company name as the key for de_ratio
  unlvr_beta[company] = round(adjusted_beta[company]/(1+(1-mtr)*de_ratio[company]),4)

print(unlvr_beta)
```

```python
median_unlvr_beta = np.median(list(unlvr_beta.values()))
print(f"Median Unlevered Beta: {median_unlvr_beta:.4f}")
```

```python

dc_ratio = {}
for company, info in companies.items():
    dc_ratio[company] = info["debt"] / (info["debt"] + info["mcap"])
    print(f"{company} Debt/Capital: {dc_ratio[company]:.4f}")


median_dc = np.median(list(dc_ratio.values()))
print(f"\nMedian Debt/Capital: {median_dc:.4f}")


target_de = median_dc / (1 - median_dc)
print(f"Target D/E: {target_de:.4f}")
```

```python
relvr_beta = round(median_unlvr_beta*(1+(1-mtr)*target_de),4)
print(f"Releverd Beta: {relvr_beta}")
```

```python
ke = rf+(relvr_beta*ERP)
print(ke)
```

```python
kd_pot = round(kd_pt*(1-mtr),6)
print(kd_pot)
```

```python
wd = target_de
we = 1-target_de

WACC = round((kd_pot*wd)+(ke*we),4)
print(f"\nThe Weighted Average Cost Of Capital for {list(companies.keys())[0]}: {WACC*100:.2f}%")
```
