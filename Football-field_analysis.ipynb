```python
import matplotlib.pyplot as plt
import pandas as pd

hul_ebitda = 15790
net_debt = -1770
shares = 234.96

ev_low = 31.7*hul_ebitda
ev_high = 41.6*hul_ebitda

ev_ebitda_low=(ev_low-net_debt)/shares
ev_ebitda_high=(ev_high-net_debt)/shares


ev_rev_low = 2088.3
ev_rev_high = 2515.9

hul_ni=15059.0
hul_eps = hul_ni/shares

low_pe = 44.8*hul_eps
high_pe = 60*hul_eps

dcf_bear_low = 523.98
dcf_bear_high = 611.71

dcf_base_low = 611.71
dcf_base_high = 755.33

dcf_bull_low = 755.33
dcf_bull_high = 1033.18

low_52wk = 2022.5
high_52wk = 2715.44

methods = [
    "52-Week Range",
    "EV/EBITDA",
    "EV/Revenue",
    "P/E",
    "DCF (Bear)",
    "DCF (Base)",
    "DCF (Bull)"
]

lows = [low_52wk,ev_ebitda_low,ev_rev_low,low_pe,dcf_bear_low,dcf_base_low,dcf_bull_low]
highs = [high_52wk,ev_ebitda_high,ev_rev_high,high_pe,dcf_bear_high,dcf_base_high,dcf_bull_high]

data_dict={
    "Methods": methods,
    "Low": lows,
    "High": highs
}

data = pd.DataFrame(data_dict)
data['Low']=data['Low'].round(1)
data['High']=data['High'].round(1)
print(data)
```

```python
data["Width"]=data["High"]-data["Low"]
data["Width"] = data["Width"].round(1)
data
```

```python
widths = [h - l for h, l in zip(highs, lows)]

fig, ax = plt.subplots(figsize=(12, 6))

ax.barh(methods, lows, color="white")

ax.barh(methods, widths, left=lows, color="steelblue", alpha=0.8)

ax.axvline(x=2317, color="red", linewidth=2,
           linestyle="--", label="Market Price ₹2,309")

ax.set_xlabel("Implied Value Per Share (₹)")
ax.set_title("HUL — Football Field Valuation Analysis")
ax.legend()
ax.grid(axis="x", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
for i, (low, high) in enumerate(zip(lows, highs)):
    ax.text(low - 30, i, f"₹{low:,.0f}",
            va="center", ha="right", fontsize=8)
    ax.text(high + 30, i, f"₹{high:,.0f}",
            va="center", ha="left", fontsize=8)

plt.tight_layout()
plt.show()
```

```python
colors = [
    "steelblue",    # 52-Week Range
    "darkorange",   # EV/EBITDA
    "darkorange",   # EV/Revenue
    "darkorange",   # P/E
    "#d9534f",      # DCF Bear
    "#e07b6b",      # DCF Base
    "#e8a89e",      # DCF Bull
]

widths = [h - l for h, l in zip(highs, lows)]

fig, ax = plt.subplots(figsize=(12, 6))

ax.barh(methods, lows, color="white")

for i, (width, low, color) in enumerate(zip(widths, lows, colors)):
    ax.barh(i, width, left=low, color=color, alpha=0.85)

for i, (low, high) in enumerate(zip(lows, highs)):
    ax.text(low - 30, i, f"₹{low:,.0f}", va="center", ha="right", fontsize=8)
    ax.text(high + 30, i, f"₹{high:,.0f}", va="center", ha="left", fontsize=8)

ax.axvline(x=2309, color="red", linewidth=2,
           linestyle="--", label="Market Price ₹2,309")

ax.set_xlabel("Implied Value Per Share (₹)")
ax.set_title("HUL — Football Field Valuation Analysis")
ax.legend()
ax.grid(axis="x", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```
